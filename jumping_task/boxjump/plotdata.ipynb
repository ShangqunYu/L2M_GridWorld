```python
lst = [1, 2, 3, 4]
N = 2
moving_average = []

def moving_average(lst, window):
    rolling_average = []
    for i in range(len(lst)):
        if i >= window-1:
            rolling_average.append(sum(lst[i-window+1:i])/window)
    return rolling_average
```
